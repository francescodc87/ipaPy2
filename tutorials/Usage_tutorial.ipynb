```python
from ipaPy2 import ipa
import pandas
import numpy
import time
import molmass
from scipy import stats
```

```python
ppmiso = 100
isoDiff = 1
ppm=2
ionisation = 1
df = pandas.read_csv('df_test_pos.csv')

DB=pandas.read_csv('DB_test_pos.csv')
DB = DB.replace(numpy.nan,None)
DB = DB.set_axis(['id', 'name', 'formula', 'inchi', 'smiles','RT','adducts',  ### just making sure the axis names are correct
                  'description','pk','MS2','reactions'],
                 axis=1, inplace=False)

#### adding MS2 data information in the annotation
#### read MS2 data
dfMS2 = pandas.read_csv("example_Dataset_MS2data.csv")
#### read MS2 database
DBMS2 = pandas.read_csv('DBMS2_test_pos.csv', low_memory=False)
DBMS2 = DBMS2.where(pandas.notnull(DBMS2), None)
```

```python
annotations = ipa.simpleIPA(df=df,ionisation=ionisation,dfMS2=dfMS2,DBMS2=DBMS2,DB=DB,adducts_file='adducts.csv',ppm=ppm,
                            delta_add=0.1,delta_bio=0.5,noits=1000,ncores=1,CSunk=0.7)
```
